# Phase diagram of hemo-ilmenite following results of Harrison 2006

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import UnivariateSpline
%config InlineBackend.figure_format='retina'
```

```python
phase_results = pd.read_csv('../data/hemo_ilm_solid_solution.csv')
phase_results
```

```python
x1 = np.linspace(0, 100, num=500)
x2 = np.linspace(0, 100, num=500)
x3 = np.linspace(55, 100, num=250)

k1 = 5
k2 = 5
k3 = 4
Tc = phase_results[1:12]['TE (K)'].tolist()+phase_results[12:]['TC (K)'].tolist()
Tn = phase_results[:20]['TN (K)'].tolist()
Te = phase_results[11:20]['TE (K)'].tolist()
Tc = [i - 273 for i in Tc]
Tn = [i - 273 for i in Tn]
Te = [i - 273 for i in Te]

Tc_smooth = np.poly1d(np.polyfit(phase_results[1:]['x'].tolist(), Tc, k1))(x1)
Tn_smooth = np.poly1d(np.polyfit(phase_results[:20]['x'].tolist(), Tn, k2))(x2)
Te_smooth = np.poly1d(np.polyfit(phase_results[11:20]['x'].tolist(), Te, k3))(x3)
# Te_smooth
```

```python
plt.figure(figsize=(8, 10))

plt.errorbar(phase_results[11:20]['x'].tolist(), Te, yerr=25, fmt='o', color = 'g')
plt.errorbar(phase_results[1:]['x'].tolist(), Tc, yerr=25, fmt='o', color='b')
plt.errorbar(phase_results[:20]['x'], Tn, yerr=25, fmt='o', color = 'k')

plt.plot(x1, Tc_smooth, color = 'b')
plt.plot(x2, Tn_smooth, color = 'k')
plt.plot(x3, Te_smooth, color = 'g')
plt.hlines(527, 0, 100, colors= 'grey', linestyles='--')
plt.ylim(0 ,1500)
plt.xlim(0, 100)
plt.ylabel('Temperature (C)')
plt.xlabel('Mol% $FeTiO_3$')
plt.show()
```
