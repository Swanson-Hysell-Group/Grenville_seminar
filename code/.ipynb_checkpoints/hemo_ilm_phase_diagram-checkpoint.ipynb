# Phase diagram of hemo-ilmenite following results of Harrison 2006

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import UnivariateSpline
%config InlineBackend.figure_format='retina'
```

```python
phase_results = pd.read_csv('../data/hemo_ilm_solid_solution.csv')
phase_results
```

```python
x1 = np.linspace(0, 100, num=500)
x2 = np.linspace(0, 100, num=500)
x3 = np.linspace(55, 100, num=250)

k1 = 5
k2 = 5
k3 = 4
Tc = phase_results[1:12]['TE (K)'].tolist()+phase_results[12:]['TC (K)'].tolist()
Tn = phase_results[:20]['TN (K)'].tolist()
Te = phase_results[11:20]['TE (K)'].tolist()
Tc = [i - 273 for i in Tc]
Tn = [i - 273 for i in Tn]
Te = [i - 273 for i in Te]

Tc_smooth = np.poly1d(np.polyfit(phase_results[1:]['x'].tolist(), Tc, k1))(x1)
Tn_smooth = np.poly1d(np.polyfit(phase_results[:20]['x'].tolist(), Tn, k2))(x2)
Te_smooth = np.poly1d(np.polyfit(phase_results[11:20]['x'].tolist(), Te, k3))(x3)
# Te_smooth
```

```python
plt.figure(figsize=(8, 10))
alpha = 0.6
plt.errorbar(phase_results[1:]['x'].tolist(), Tc, yerr=25, fmt='o', color='C0', alpha=alpha)
plt.errorbar(phase_results[:20]['x'], Tn, yerr=25, fmt='o', color = 'C1', alpha=alpha)
plt.errorbar(phase_results[11:20]['x'].tolist(), Te, yerr=25, fmt='o', color = 'C2', alpha=alpha)

plt.plot(x1, Tc_smooth, color = 'C0', alpha=alpha)
plt.plot(x2, Tn_smooth, color = 'C1', alpha=alpha)
plt.plot(x3, Te_smooth, color = 'C2', alpha=alpha)
plt.hlines(527, 0, 100, colors= 'grey', linestyles='--')
plt.text(40, 1100, r'PM R$\bar{3}$c', fontsize=14)
plt.text(80, 900, r'PM R$\bar{3}$', fontsize=14)
plt.text(50, 600, r'PM R$\bar{3}$c+PM R$\bar{3}$', fontsize=12)
plt.text(40, 150, r'AF R$\bar{3}$c+PM R$\bar{3}$', fontsize=14)
plt.annotate(r'AF R$\bar{3}$c', (5, 550), (15, 650), arrowprops=dict(arrowstyle="->"), fontsize=14)

plt.ylim(100 ,1400)
plt.xlim(0, 100)
plt.yticks([100, 300, 527, 700, 900, 1100, 1300])
plt.ylabel('Temperature ($^\circ$C)', fontsize=14)
plt.xlabel('Mol% FeTiO$_3$', fontsize=14)
plt.savefig('../figure/hemo_ilm_phase_diagram.pdf')
plt.show()
```
